```python
%sql
select e.emp_name, sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name;
```

```python
%sql
select e.emp_name, count(s.sales_id) as total_sales_count
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name;
```

```python
%sql
select product, sum(amount) as total_sales
from sales
group by product;
```

```python
%sql
select product, avg(amount) as avg_sales
from sales
group by product;
```

```python
%sql
select e.emp_name, count(s.sales_id) as total_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name
having count(s.sales_id) > 2;
```

```python
%sql
select e.emp_name, count(s.sales_id) as total_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name
having count(s.sales_id) > 2;
```

```python
%sql
select e.emp_name, count(distinct s.product) as unique_products
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name;
```

```python
%sql
select e.emp_name, max(s.amount) as highest_sale
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name;
```

```python
%sql
select product, sum(amount) as total_sales
from sales
group by product
having sum(amount) > 50000;
```

```python
%sql
select e.department, avg(s.amount) as avg_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.department
having avg(s.amount) = (
    select max(avg_sales)
    from (
        select avg(s.amount) as avg_sales
        from employee e
        join sales s
        on e.emp_id = s.emp_id
        group by e.department
    ) t
);
```

```python
%sql
select e.department, sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.department
having sum(s.amount) = (
    select max(total_sales)
    from (
        select sum(s.amount) as total_sales
        from employee e
        join sales s
        on e.emp_id = s.emp_id
        group by e.department
    ) t
);
```

```python
%sql
select e.emp_name, sum(s.amount) as total_sales
from employee e
join sales s
on e.emp_id = s.emp_id
group by e.emp_name
order by total_sales desc
limit 3;
```
